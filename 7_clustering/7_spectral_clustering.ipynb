<center><h2>When k-means is not enough</h2></center>
<br>
<center><img src="../images/k-means+vs.+Spectral+Clustering.png" width="75%"/></center>

K-means works best with spherical data. Often data is not spherical data.

We want to cluster because data is close but not spherical.

<center><h2>The goal of spectral clustering</h2></center>

To cluster data that is connected but not necessarily compact or clustered within convex boundaries.
<br>
<center><img src="../images/compact.png" width="75%"/></center>

<center><h2>Graph Clustering</h2></center>

Graphs have nodes and edges. Often the edges have weights.
<br>
<center><img src="../images/graphs.png" width="75%"/></center>

<center><h2>Spectral Clustering Steps</h2></center>

1. Build similarity graph between data objects. Typically, pick a distance metric and calculate for each set of pairs.
1. Pick number of clusters (k).
1. Compute the first k eigenvectors of its Laplacian matrix to embedded lower dimensional space.
1. Run clustering (e.g., k-means) on these features to separate objects into k classes.

<center><h2>How to build a Laplacian matrix</h2></center>

A - Adjacency matrix

<center><img src="../images/adjenciy matrix.png" width="45%"/></center>

D - diagonal matrix of degrees

<center><img src="../images/d.png" width="15%"/></center>


Laplacian matrix is the the difference of the diagonal matrix of its vertex degrees and its its adjacency matrix.

$$L = D - A$$

<center><img src="../images/lap.png" width="35%"/></center>

The find eigenvalues and eigenvector of Laplacian matrix 

$$Lx=λx$$

Sort x for given λ (typically λ<sub>2</sub>)

| Rank | Value |  
|:-------:|:------:|
| 1 | .8 |
| 2 | .45 |
| 3 | -0.4 |
| 4 | -.55 |

There is often a natural break point(s) for clusters. 

For example, zero is a common break point. Thus, Cluster A are positive points and Cluster B are negative points.

[Source](https://www.youtube.com/watch?v=zkgm0i77jQ8&ab_channel=OmarSobh)

<center><h2>What is Spectral Clustering?</h2></center>

Performs a low-dimension embedding of the affinity matrix between samples. Followed by clustering of the components of the eigenvectors in the low dimensional space.

Use the spectrum (eigenvalues) of the Laplacian (similarity) matrix of the data to perform dimensionality reduction before clustering in fewer dimensions

<center><h2> Why does Spectral Clustering work?</h2></center>

Data are projected into a lower-dimensional space (the spectral/eigenvector domain) where they are easily separable, say using k-means.

<center><img src="../images/projection.png" width="75%"/></center>

<center><h2>The primary hyperparameters of spectral clustering</h2></center>

1. The number of clusters
1. Affinity (aka, distance metric)

> When calling fit, an affinity matrix is constructed using either kernel function such the Gaussian (aka RBF) kernel of the euclidean distanced d(X, X):

<center><h2>Spectral Clustering in scikit-learn</h2></center>

```python
reset -fs
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# Create toy dataa
import numpy as np
X = np.array([[1, 1], [2, 1], [1, 0],
              [4, 7], [3, 5], [3, 6]])
```

```python
import matplotlib.pyplot as plt
palette = "Dark2"
%matplotlib inline
```

```python
plt.scatter(X[:,0], X[:,1]);
```

```python
from sklearn.cluster import SpectralClustering
```

Let's read the documentation

https://scikit-learn.org/stable/modules/generated/sklearn.cluster.SpectralClustering.html

```python
sp = SpectralClustering(n_clusters=2)
sp.fit(X)
```

```python
# Plot labels
plt.scatter(X[:,0], X[:,1], c=sp.labels_);
```

```python
# A slightly more complex example
X_complex = np.load("../data/slighlty_complex_data.npy") 
plt.scatter(X_complex[:, 0], X_complex[:, 1]);
```

```python
sp = SpectralClustering(n_clusters=2)
sp.fit(X_complex)
plt.scatter(X_complex[:,0], X_complex[:,1], c=sp.labels_);
```

```python
# Change hyperparmaters to get a better fit
# The default is ‘rbf’ : construct the affinity matrix using a radial basis function (RBF) kernel.
# That Guassian type assumption makes less sense for this data
# Switch to ‘nearest_neighbors’ : construct the affinity matrix by computing a graph of nearest neighbors.
sp = SpectralClustering(n_clusters=2,
                        affinity='nearest_neighbors')
sp.fit(X_complex)
plt.scatter(X_complex[:,0], X_complex[:,1], c=sp.labels_);
```

<center><h2>Connection to Graph Theory</h2></center>

We are casting our data point cloud to a graph. However, sometimes you'll have native graph data. Social networks or computer networks are common graphs in data science.

These spectral methods can be used to find communities in graphs, formally called Community Detection. For example, what groups on Twitter are talking to each other but not to other groups?

Spectral Clustering can be used for Community Detection.

<center><h2>Takeaways</h2></center>

- Spectral clustering (SC) looks the weighted graph connections of data points.
- SC projects them into a lower dimensional space by finding the eigenvalues of the Laplacian matrix. The clusters happens in this feature space.
- SC can find useful clusters of non-spherical data.
